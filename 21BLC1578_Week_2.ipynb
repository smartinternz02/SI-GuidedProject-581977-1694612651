```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
print(sns.get_dataset_names())
```

```python
df=sns.load_dataset('car_crashes')
```

```python
df
```

```python
sns.scatterplot(x="total",y="speeding",data=df)
#inference: This is a scatter plot representing the relationship between the total and speeding
```

```python
sns.lineplot(x="alcohol",y="not_distracted",data=df)
#inference: #inference: This is a line plot representing the relationship between alcohol and not_distracted
```

```python
sns.distplot(df["alcohol"])
#inference: #inference: This is a distribution plot representing alcohol
```

```python
sns.relplot(x="total",y="not_distracted",data=df)
#inference: This is a relational plot representing the relationship between the total and not_distracted
```

```python
sns.barplot(x="ins_premium",y="ins_losses",data=df,ci=None)
#inference: This is a bar plot representing the relationship between ins_premium and ins_losses
```

```python
sns.countplot(x="not_distracted",data=df)
#inference: This is a count plot representing not_distracted
```

```python
sns.jointplot(x="total",y="speeding",data=df)
#inference: This is a joint plot representing the relationship between the total and speeding
```

```python
sns.boxplot(x="total",y="speeding",data=df)
#inference: This is a box plot representing the relationship between the total and speeding
```
